```python
from tqdm import tqdm
import pandas as pd
import numpy as np
import librosa 
import librosa.display
import pylab
import os
import pickle
```

```python
# 이진분류 데이터셋 
#trainData.pickle파일 불러오기
with open('../22.03.30 V1 데이터/pickle/T1_rawTrain.pickle', 'rb') as f:
    trainData = pickle.load(f) 
```

```python
Current_mel = []
VibX_mel = []
VibY_mel = []
VibZ_mel = []
```

```python
trainData
```

```python
# TrainData 2D Transform
Current_figpath = '../22.03.30 V1 데이터/Mel spectrogram/Train_Current'
VibX_figpath = '../22.03.30 V1 데이터/Mel spectrogram/Train_VibX'
VibY_figpath = '../22.03.30 V1 데이터/Mel spectrogram/Train_VibY'
VibZ_figpath = '../22.03.30 V1 데이터/Mel spectrogram/Train_VibZ'

def feature_extractor(row):
    name     = "T"+str(row[6])+"B"+str(row[7])+"TC"+str(row[4])
#     try:  
    audio1 = np.array(row[0])
    audio2 = np.array(row[1])
    audio3 = np.array(row[2])
    audio4 = np.array(row[3])

    # Current
    melC = librosa.feature.melspectrogram(y=audio1, sr=1000)
    s_db_C     = librosa.power_to_db(melC, ref=np.max)
    pylab.axis('off') # no axis
    pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[])
    librosa.display.specshow(s_db_C, sr = 1000)
    Current_path = os.path.join(Current_figpath,name+'.png')
    pylab.savefig(Current_path, bbox_inches=None, pad_inches=0)
    pylab.close()  

    # VibX
    melX = librosa.feature.melspectrogram(y=audio2, sr=1000)
    s_db_X     = librosa.power_to_db(melX, ref=np.max)
    pylab.axis('off') # no axis
    pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[])
    librosa.display.specshow(s_db_X, sr = 1000)
    VibX_path = os.path.join(VibX_figpath,name+'.png')
    pylab.savefig(VibX_path, bbox_inches=None, pad_inches=0)
    pylab.close() 

    # VibY
    melY = librosa.feature.melspectrogram(y=audio3, sr=1000)
    s_db_Y     = librosa.power_to_db(melY, ref=np.max)
    pylab.axis('off') # no axis
    pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[])
    librosa.display.specshow(s_db_Y, sr = 1000)
    VibY_path = os.path.join(VibY_figpath,name+'.png')
    pylab.savefig(VibY_path, bbox_inches=None, pad_inches=0)
    pylab.close() 

    # VibZ
    melZ = librosa.feature.melspectrogram(y=audio4, sr=1000)
    s_db_Z     = librosa.power_to_db(melZ, ref=np.max)
    pylab.axis('off') # no axis
    pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[])
    librosa.display.specshow(s_db_Z, sr = 1000)
    VibZ_path = os.path.join(VibZ_figpath,name+'.png')
    pylab.savefig(VibZ_path, bbox_inches=None, pad_inches=0)
    pylab.close() 


#     except:
#         print('file cannot open')
#         return None
    return Current_path, VibX_path, VibY_path, VibZ_path
```

```python
#저장후 이미지 경로 리스트에 저장
for row in tqdm(trainData.values):
    Current_path, VibX_path, VibY_path, VibZ_path = feature_extractor(row)
    Current_mel.append(Current_path)
    VibX_mel.append(VibX_path)
    VibY_mel.append(VibY_path)
    VibZ_mel.append(VibZ_path)
```

```python
print(len(Current_mel))
```

```python
print(len(VibX_mel))
```

```python
#None 아닌 데이터만 추출 및 label 1,0으로 저장
isnone = lambda x: x != None
label  = lambda xx: 1 if xx == True else 0
cast_x = list(map(isnone,Current_mel)) # check missing parts 
data_y = list(map(label,trainData['Label'])) # label: Boolean -> binary
data_x1 = [Current_mel[i] for i in range(len(Current_mel)) if cast_x[i] == True] # mel-sp = x
data_x2 = [VibX_mel[i] for i in range(len(VibX_mel)) if cast_x[i] == True] # mel-sp = x
data_x3 = [VibY_mel[i] for i in range(len(VibY_mel)) if cast_x[i] == True] # mel-sp = x
data_x4 = [VibZ_mel[i] for i in range(len(VibZ_mel)) if cast_x[i] == True] # mel-sp = x
data_x5 = [trainData['Current'][i] for i in range(len(Current_mel)) if cast_x[i] == True] # mel-sp = x
data_x6 = [trainData['VibX'][i] for i in range(len(Current_mel)) if cast_x[i] == True] # mel-sp = x
data_x7 = [trainData['VibY'][i] for i in range(len(Current_mel)) if cast_x[i] == True] # mel-sp = x
data_x8 = [trainData['VibZ'][i] for i in range(len(Current_mel)) if cast_x[i] == True] # mel-sp = x
data_y = [data_y[i] for i in range(len(Current_mel)) if cast_x[i] == True] # label = y 
```

```python
indices = np.arange(len(data_x1))
DATA    = {}
np.random.shuffle(indices)
DATA['MEL_C'] = [data_x1[i] for i in indices]
DATA['MEL_X'] = [data_x2[i] for i in indices]
DATA['MEL_Y'] = [data_x3[i] for i in indices]
DATA['MEL_Z'] = [data_x4[i] for i in indices]
DATA['1D_C'] = [data_x5[i] for i in indices]
DATA['1D_X'] = [data_x6[i] for i in indices]
DATA['1D_Y'] = [data_x7[i] for i in indices]
DATA['1D_Z'] = [data_x8[i] for i in indices]
DATA['LABEL'] = [data_y[i] for i in indices]
```
