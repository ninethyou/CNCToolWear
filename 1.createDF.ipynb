```python
import os
import glob
import pandas as pd
import numpy as np
import pickle
from tqdm import tqdm
from pandas import DataFrame as df
import matplotlib.pyplot as plt
import pickle
```

```python
# load data


files = glob.glob("../22.03.30 V1 데이터/unzipped/*/2022*",recursive = True)
files.sort()

data = pd.DataFrame(columns = ['Current',"VibX","VibY",'VibZ',"ToolCount", "ProductCount","ToolNumber"])
for file in tqdm(files,desc='dirs'):
    try:
        temp = pd.read_csv(file,names=['Current',"VibX","VibY",'VibZ',"ToolCount", "ProductCount","ToolNumber"])


        condition = (temp["ToolNumber"] == 1)
        cd_tool_temp = temp[condition]

        cd_Current_temp = cd_tool_temp["Current"]          
        cd_Current_temp_np = np.array(cd_Current_temp)
        
        cd_VibX_temp = cd_tool_temp["VibX"]
        cd_VibX_temp_np = np.array(cd_VibX_temp)
        
        cd_VibY_temp = cd_tool_temp["VibY"]
        cd_VibY_temp_np = np.array(cd_VibY_temp)
        
        cd_VibZ_temp= cd_tool_temp["VibZ"]
        cd_VibZ_temp_np = np.array(cd_VibZ_temp)
               
        new_data = {
                            "Current" : cd_Current_temp_np, 
                            "VibX" : cd_VibX_temp_np, 
                            "VibY" : cd_VibY_temp_np ,
                            "VibZ": cd_VibZ_temp_np  ,
                            "ToolCount": cd_tool_temp["ToolCount"].unique()[0],
                            "ProductCount" : cd_tool_temp["ProductCount"].unique()[0],
                            "ToolNumber": cd_tool_temp["ToolNumber"].unique()[0],

                 } 
        data= data.append(new_data,ignore_index= True)

    except:
           continue 
        
```

```python
plt.plot(np.arange(len(data['Current'][0])),data['Current'][0])
plt.show()
```
