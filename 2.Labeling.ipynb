```python
import os
import glob
import pandas as pd
import numpy as np
import pickle
from tqdm import tqdm
```

```python
with open("../22.03.30 V1 데이터/pickle/T1_rawData.pickle", 'rb') as f:
    data = pickle.load(f)
```

```python
data
```

```python
toolcount_list = data['ToolCount'].to_list()
bites = []

bite_number = 1
current_tc = 0
previous_tc = -1

for tc in toolcount_list:
    current_tc = tc
    
    if current_tc >= previous_tc:
        bites.append(bite_number)
        
    else:
        bite_number += 1
        bites.append(bite_number)
        
    previous_tc = current_tc

data['biteCount'] = bites

del toolcount_list
```

```python
df = pd.DataFrame(columns = ['Current','VibX','VibY','VibZ','ToolCount','ProductCount','ToolNumber','biteCount'])

for i in range(1,max(data['biteCount'])):
    tmp = data[data['biteCount'] == i]
    print(tmp['ToolCount'])
    tmp=tmp.reset_index(drop=True)
    print(tmp)
    if (len(tmp['ToolCount']) >= 200)&(tmp['ToolCount'][0]==1):
        df = df.append(tmp,ignore_index= True)
print(df)
```

```python
falseData = pd.DataFrame(columns = ['Current','VibX','VibY','VibZ','ToolCount','ProductCount','ToolNumber','biteCount','Label'])
trueData = pd.DataFrame(columns = ['Current','VibX','VibY','VibZ','ToolCount','ProductCount','ToolNumber','biteCount','Label'])

for i in df['biteCount'].unique():
    falseData = falseData.append(df[df["biteCount"] == i][:20],ignore_index= True)
    trueData = trueData.append(df[df["biteCount"] == i][-20:],ignore_index = True)

falseData['Label'] = False
trueData['Label']=True
```

```python
boundary = df['biteCount'].unique()[round(len(df['biteCount'].unique())*0.9)]

trainData = (falseData[falseData["biteCount"] < boundary])
trainData =  trainData.append(trueData[trueData["biteCount"] < boundary],ignore_index = True)

valData = (falseData[falseData["biteCount"] >= boundary])
valData =  valData.append(trueData[trueData["biteCount"] >= boundary],ignore_index = True)
```
